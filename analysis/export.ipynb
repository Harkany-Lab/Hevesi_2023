```python
# Combine all gene sets
all_genes = list(set(nmr + npr + npep + neurotrans))

all_genes
```

```python
# Function to process and save data
def process_and_save_data(adata, dataset_name):
    # Subset genes
    selected_genes = [gene for gene in all_genes if gene in adata.var_names]
    adata = adata[:, selected_genes]

    # 1. Normalized expression matrix
    normalized = pd.DataFrame(
        adata.layers["RAW"].toarray(), index=adata.obs_names, columns=adata.var_names
    )
    normalized.to_csv(f"output/tables/{dataset_name}_normalized.csv")

    # 2. Scaled expression matrix
    scaled = pd.DataFrame(
        adata.X.toarray(), index=adata.obs_names, columns=adata.var_names
    )
    scaled.to_csv(f"output/tables/{dataset_name}_scaled.csv")

    # 3. Average normalized expression per cluster
    avg_normalized = normalized.groupby(adata.obs["k_tree"]).mean()
    avg_normalized.to_csv(
        f"output/tables/{dataset_name}_avg_normalized_per_cluster.csv"
    )
```

```python
# Create output directory
Path("output/tables").mkdir(parents=True, exist_ok=True)

# Process and save data for both datasets
process_and_save_data(thp7_data, "THP7")
process_and_save_data(pr5p7_data, "Pr5P7")

print("Data processing and saving completed.")
```
